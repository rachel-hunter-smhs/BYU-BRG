```
#BASIC CHALLENGE

#Here is your data
from matplotlib import pyplot as plt
import numpy as np
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

plt.plot(t,s,"c", linewidth=0.4)
plt.grid(color='r', linestyle=':', linewidth=.4)
plt.show()
```

```
#INTERMEDIATE CHALLENGE

#Here is your data
N = 5
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
labels = ['G1', 'G2', 'G3', 'G4', 'G5']

plt.bar(ind, menMeans, color="blue", alpha=0.4, yerr=menStd)
plt.bar(ind, womenMeans,bottom=menMeans, color="orange", alpha=0.4, yerr=womenStd)
plt.ylabel('scores')
plt.title('scores by group and gender')
#xticklabels(labels)


plt.show()

```
